```julia
plt1 = plot(sin, label="sin")
plt2 = plot(cos, label="cos")
plt3 = plot(exp, label="exp")
plt4 = plot(abs, label="abs")

plot(plt1, plt2, plt3, plt4, layout = (2, 2))
```

```julia
plot(
    plot(sin, label="sin"),
    plot(cos, label="cos"),
    plot(exp, label="exp"),
    plot(abs, label="abs"),
    layout = (2, 2)
)
```

## グラフを重ねる

```julia
plt = plot(sin, label = "sin")
plot!(plt, cos, label = "cos")
```

```julia
X = [i for i ∈ 0:10]
Y = sin.(X)

plt = plot(0:0.1:10, sin, label="function")
scatter!(plt, X, Y, label="array")
```

```julia
plot(x -> x * sin(x^2), label="")
```

```julia
f(x) = exp(-x^2/2) / sqrt(2 * π)
X = randn(1000)
Y = [-0.01 - 0.0001 * i for i ∈ 1:1000]

plt = histogram(X, bins=range(-5, 5, step=0.2), ylims=(-0.14,), normed=true, label="histogram", fa=0.3, fc=1, lc="#FFFFFF")
plot!(plt, f, lw=2, lc=:black, label="Exact PDF")
```

```julia
f(x) = ℯ^((-x^2/2)) / sqrt(2 * π)
X = randn(1000)
Y = [-0.01 - 0.00005 * i for i ∈ 1:1000]

plt = histogram(X, bins=range(-5, 5, step=0.2), normed=true, ylims=(-0.7,), fa=0.3, fc=1)
scatter!(plt, X, Y, ms = 1, msw=0, mc="#3261Ab", lable="Sample")
plot!(plt, f, lw=2, lc=:black, label="Exact PDF")
```
